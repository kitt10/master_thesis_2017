```python
from shelve import open as open_shelve
from kitt_net import FeedForwardNet
from numpy import zeros

dataset = open_shelve('../examples/speech/dataset_speech_5K_bs3_cs0.ds', 'c')
net = FeedForwardNet(hidden=[50], tf_name='Sigmoid')
net.fit(x=dataset['x'], y=dataset['y'], x_val=dataset['x_val'], y_val=dataset['y_val'], n_epoch=1,
       learning_rate=0.1, batch_size=10)
net.load('../examples/speech/net_speech_5K_bs3_cs0_s50.net')
net.saliency = (zeros(1),)

net.prune(req_acc=0.35, req_err=0.05, n_epoch=3, levels=(75, 50, 40, 30, 25, 20, 15, 10, 7, 5, 3, 2, 1, 0), 
          verbose_learning=True, max_steps=30)
```

```python
net.dump('../examples/speech/net_speech_5K_bs3_cs0_s50_pruned.net')
```
